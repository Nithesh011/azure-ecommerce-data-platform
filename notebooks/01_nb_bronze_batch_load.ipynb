CELL 1 — Define paths

```python
storage = "your-storage-name"

raw_products = f"abfss://datalake@{storage}.dfs.core.windows.net/raw/fakestore/products/"
raw_users    = f"abfss://datalake@{storage}.dfs.core.windows.net/raw/fakestore/users/"

bronze_products = f"abfss://datalake@{storage}.dfs.core.windows.net/bronze/fakestore/products/"
bronze_users    = f"abfss://datalake@{storage}.dfs.core.windows.net/bronze/fakestore/users/"
```

CELL 2 — Read RAW data and Write to BRONZE

```python
df_products_raw = spark.read.json(raw_products)
df_users_raw = spark.read.json(raw_users)
df_products_raw.write.mode("overwrite").parquet(bronze_products)
df_users_raw.write.mode("overwrite").parquet(bronze_users)
print("Bronze batch load completed!")
```
